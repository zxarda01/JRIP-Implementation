```
# Preprocess Data
import pandas as pd
df = pd.read_csv('/content/mushrooms.csv')
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=.33, random_state=42)
val, test = train_test_split(test, test_size=.99, random_state=42)

df.info()
```

```
# train classifier
import wittgenstein as lw
ripper_clf = lw.RIPPER(random_state=42) # Or irep_clf = lw.IREP() to build a model using IREP
ripper_clf.fit(train, class_feat='type',  pos_class='p') # Or pass X and y data to .fit

# processing Test Data 
x_test = test.drop('type', axis=1)
y_test = test['type']

# performance Metic
from sklearn.metrics import precision_score, recall_score
precision = ripper_clf.score(x_test, y_test, precision_score)
recall = ripper_clf.score(x_test, y_test, recall_score)
cond_count = ripper_clf.ruleset_.count_conds()
print(f'precision: {precision} recall: {recall} conds: {cond_count}')
```

```
# Prediction 
#print(val)
x_val = test.drop('type', axis=1)
y_val = test['type']
ripper_clf.predict(x_val)
```

```
# explaination
ripper_clf.predict(x_val, give_reasons=True)
```
